## Estrutura de repetição

Faça um programa que peça uma nota, entre zero e dez. Mostre uma mensagem caso o valor seja inválido e continue pedindo até que o usuário informe um valor válido.

```python
while True:
    try:
        nota = input("Forneça uma nota de 0 a 10, por favor: ")
        nota = float(nota) # Converte para float

        if 0 <= nota <= 10:
            print(f"Correto, seu número é {nota}")
            return True # Para o loop e a execução da função
        else:
            # Cai aqui se nota for 11, -1, etc. O loop continua.
            print("Número errado, tente novamente")
            
    except ValueError:
        # Cai aqui se a entrada não puder virar float (ex: 'abc'). O loop continua.
        print("Caracter invalido, tente novamente")
        

```
